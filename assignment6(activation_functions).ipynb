```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Input range
x = np.linspace(-10, 10, 200)

# 1 Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 2 Tanh
def tanh(x):
    return np.tanh(x)

# 3 ReLU
def relu(x):
    return np.maximum(0, x)

# 4 Leaky ReLU
def leaky_relu(x):
    return np.where(x > 0, x, 0.01 * x)

# 5 ELU
def elu(x, alpha=1):
    return np.where(x > 0, x, alpha*(np.exp(x)-1))

# 6 Softmax
def softmax(x):
    e = np.exp(x - np.max(x))
    return e / e.sum()

# 7 Softplus
def softplus(x):
    return np.log(1 + np.exp(x))

# 8 Binary Step
def binary_step(x):
    return np.where(x >= 0, 1, 0)

# 9 Piecewise Linear
def piecewise_linear(x):
    return np.where(x < -1, 0, np.where(x <= 1, x, 1))

# Plotting all functions
functions = [
    ("Sigmoid", sigmoid(x)),
    ("Tanh", tanh(x)),
    ("ReLU", relu(x)),
    ("Leaky ReLU", leaky_relu(x)),
    ("ELU", elu(x)),
    ("Softmax", softmax(x)),
    ("Softplus", softplus(x)),
    ("Binary Step", binary_step(x)),
    ("Piecewise Linear", piecewise_linear(x))
]

plt.figure(figsize=(12,10))

for i, (name, y) in enumerate(functions):
    plt.subplot(3,3,i+1)
    plt.plot(x, y)
    plt.title(name)
    plt.grid()

plt.tight_layout()
plt.show()
```
